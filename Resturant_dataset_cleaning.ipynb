```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r"C:\Users\matur\Downloads\Restaurant_Raw_Dataset.xlsx")
```

```python
df
```

```python
# Remove duplicates
df = df.drop_duplicates()
```

```python
df.loc[:, "Order_Date"] = pd.to_datetime(
    df["Order_Date"],
    errors="coerce",
    dayfirst=True
)
```

```python
df
```

```python
df = df.dropna(subset=["Order_Date"])
```

```python
# Convert numeric columns
numeric_cols = [
    "Quantity",
    "Unit_Price (₹)",
    "Total_Sales (₹)",
    "Cost_Price (₹)",
    "Profit (₹)",
    "Profit_Margin (%)",
    "Customer_Rating"
]
# If df is a filtered DataFrame
df = df.copy()
# Convert columns to numeric
for col in numeric_cols:
    df.loc[:, col] = pd.to_numeric(df[col], errors="coerce")
```

```python
# Remove invalid values
df = df[df["Quantity"] > 0]
df = df[df["Unit_Price (₹)"] > 0]
df = df[df["Total_Sales (₹)"] > 0]
df = df[df["Cost_Price (₹)"] > 0]
df = df[df["Profit (₹)"] >= 0]
df = df[df["Profit_Margin (%)"] <= 100]
```

```python
# Payment validation
valid_payment = ["Cash","UPI","Credit Card","Debit Card"]
df = df[df["Payment_Method"].isin(valid_payment)]
```

```python
# Order type validation
valid_order = ["Dine-In","Takeaway","Delivery"]
df = df[df["Order_Type"].isin(valid_order)]
```

```python
# Age validation
valid_age = ["18–24","25–34","35–44","45–54","55+"]
df = df[df["Customer_Age_Group"].isin(valid_age)]
```

```python
# Valid Order ID
df = df[df["Order_ID"].str.match(r"^ORD\d+$")]
```

```python
# Remove outliers
Q1 = df["Total_Sales (₹)"].quantile(0.25)
Q3 = df["Total_Sales (₹)"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[
    (df["Total_Sales (₹)"] >= lower) &
    (df["Total_Sales (₹)"] <= upper)
]
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
# Save cleaned dataset
df.to_excel("Restaurant_Clean_Dataset.xlsx", index=False)
```
